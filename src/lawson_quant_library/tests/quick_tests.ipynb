```python
from lawson_quant_library.parameter.ir_curve import IRCurve
from lawson_quant_library.util import to_ql_date, Calendar, parse_tenor

rate = 0.05
cal = Calendar("US:NYSE", day_count="ACT365F")
cal._ql_calendar()
rate_curve = IRCurve(rate, cal)

ref_date = to_ql_date("2026-02-07")

quotes = {"1W": 0.01, "1M": 0.02, "2Y": 0.10}

tenors = [parse_tenor(t) for t in quotes.keys()]


rate_curve.set_deposit_quotes(quotes=quotes, reference_date=ref_date)

rate_curve.table(tenors=tenors, reference_date=ref_date)
```
