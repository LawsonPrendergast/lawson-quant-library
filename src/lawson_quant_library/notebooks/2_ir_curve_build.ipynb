```python
from lawson_quant_library.parameter import IRCurve
```

```python
quotes = {
    "1M": 0.0500,
    "3M": 0.0520,
    "6M": 0.0530,
    "1Y": 0.0550,
}
```

```python
curve = IRCurve.from_deposit_quotes(
    quotes,
    reference_date="2026-01-06",
    name="USD_DEPOSITS_TEST",
)
```

```python
curve.table(["1M", "3M", "6M", "1Y"])
```

```python
curve.discount("2026-07-06")
```

```python
curve.zero_rate("2027-01-06")
```

```python
curve.forward_rate("2026-01-06", "2026-07-06")
```
